```python
import sys
import seaborn as sns
import pandas as pd
import snap
import networkx as nx
from tqdm import tqdm
```

To use run the next command please download the "higgs-social_network.edgelist" from the link provided in the Data folder

```python
followers = nx.read_edgelist(path="Data/higgs-social_network.edgelist", nodetype=int, create_using=nx.DiGraph())
```

```python
retweet_reduced = nx.read_edgelist(path="Data/higgs-retweet_network_reduced.edgelist", nodetype=int, data=(('weight',int),), create_using=nx.DiGraph())
```

```python
paths = []
for src, dst, w in tqdm(retweet_reduced.edges.data()):
  try:
    path = nx.shortest_path(followers, src, dst)
    paths.append((path, w['weight']))
  except:
    pass

#paths[:5]
```

```python
edges = {}
for path, w in tqdm(paths):
    for i in range(len(path) - 1):
        if (path[i], path[i + 1]) in edges:
          print(path[i])
          edges[(path[i], path[i + 1])] += w
        else:
          edges[(path[i], path[i + 1])] = w
    
#edges
```

```python
final_edges = []

for key, value in tqdm(edges.items()):
    temp = (*key, {'weights': value})
    final_edges.append(temp)

#final_edges
```

```python
spreading_reduced = nx.DiGraph()
spreading_reduced.add_edges_from(final_edges)
spreading_reduced.edges.data()
```

```python
import numpy as np
def NormalizeData(data):
    return (data - np.min(data)) / (np.max(data) - np.min(data))
def NormalizeNormalData(data):
    return data / np.linalg.norm(data)
```

```python
weights = {}

for (src, dst), value in tqdm(edges.items()):
    weights[(src, dst)] = value

norm = NormalizeData(list(weights.values()))

for i, ((src, dst), value) in enumerate(tqdm(edges.items())):
    weights[(src, dst)] = norm[i] if norm[i] > 0.1 else 0.1
weights
```

```python
edge_list = []

for (src, dst), value in tqdm(edges.items()):
    temp = f"{src} {dst} {weights[(src, dst)]}\n"
    edge_list.append(temp)

edge_list[:5]
```

```python
with open("data/spreading_reduced.edgelist", "w") as file_:
    # Writing data to a file
    file_.writelines(edge_list)
```
